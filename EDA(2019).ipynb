# EDA

### Categorical variables

```python
# FUEL CATEGORY
sns.countplot(x='FUEL CATEGORY', data=df)
plt.tight_layout()  
plt.savefig("Fuel Category Distribution.png", dpi=300)  
plt.title("Fuel Category Distribution")
```

```python
# VEHICLE CLASS CATEGORY - Pie Chart
df['VEHICLE CLASS CATEGORY'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.tight_layout()  
plt.savefig("Vehicle class category Breakdown.png", dpi=300) 
plt.title("Vehicle Class Breakdown")
```

```python
# District
print("Total unique districts:", df['DISTRICT'].nunique())
```

```python
all_districts_counts = df['DISTRICT'].value_counts()
```

```python
plt.figure(figsize=(10, 15)) # Adjust height as needed to fit all bars
plt.tight_layout()  
plt.savefig("All District Counts.png", dpi=300) 
sns.barplot(x=all_districts_counts.values, y=all_districts_counts.index, palette='viridis')
plt.title("All Districts Count")
```

```python
print("Total unique Brands:", df['BRAND'].nunique())
```

```python
# TOP 20 BRANDS
top_brands = df['BRAND'].value_counts().head(20)
top_brands.plot(kind='bar')
plt.tight_layout()  
plt.savefig("Top 20 Brands.png", dpi=300) 
plt.title("Top 20 Brands")
```

```python
# FUEL VS BRAND (Top 10 Brands)
top_brands_df = df[df['BRAND'].isin(top_brands.index)]
pd.crosstab(top_brands_df['BRAND'], top_brands_df['FUEL CATEGORY']).plot(kind='bar', stacked=True)
plt.tight_layout()  
plt.savefig("Fuel VS Brand.png", dpi=300) 
plt.title ("Fuel VS Brand TOP 10")
```

```python
# ZONE-WISE ANALYSIS
sns.countplot(y='ZONE', data=df, order=df['ZONE'].value_counts().index)
plt.tight_layout()  
plt.savefig("Vehicle Registrations by Zone.png", dpi=300) 
plt.title("Vehicle Registrations by Zone")
```

### Numerical Variables

```python
# SEAT CAPACITY
sns.histplot(df['SEAT CAPACITY'], bins=20, kde=True)
plt.tight_layout()  
plt.savefig("Seat Capacity Distribution.png", dpi=300) 
plt.title("Seat Capacity Distribution")
```

# EV FOCUSED

```python
# EV vs All Others
```

```python
df['EV_vs_Others'] = df['FUEL CATEGORY'].apply(lambda x: 'EV' if x == 'EV' else 'Non-EV')
ev_counts = df['EV_vs_Others'].value_counts()
plt.figure(figsize=(6,6))
ev_counts.plot.pie(autopct='%1.1f%%', startangle=90, colors=['lightgreen', 'lightcoral'])
plt.title("EV vs Non-EV Share")
plt.ylabel('')  
plt.tight_layout() 
plt.savefig("EV vs Non-EV Share.png", dpi=300) 
plt.show()
```

```python
# Districts with High EV Adoption
```

```python
ev_district = df[df['FUEL CATEGORY'] == 'EV']['DISTRICT'].value_counts().head(10)
ev_district.plot(kind='bar', color='green')
plt.tight_layout()  
plt.savefig("Top 10 Districts with Highest EV Registrations.png", dpi=300) 
plt.title("Top 10 Districts with Highest EV Registrations")
```

```python
# Fuel category by zone
```

```python
fuel_zone = pd.crosstab(df['ZONE'], df['FUEL CATEGORY'])
fuel_zone.plot(kind='bar', stacked=True, figsize=(10,6), colormap='Set3')
plt.tight_layout()  
plt.savefig("Fuel Category Distribution by Zone.png", dpi=300) 
plt.title("Fuel Category Distribution by Zone")
```

```python
# TOP EV BRANDS
```

```python
ev_brands = df[df['FUEL CATEGORY'] == 'EV']['BRAND'].value_counts().head(10)
sns.barplot(x=ev_brands.values, y=ev_brands.index, palette='crest')
plt.tight_layout()  
plt.savefig("TOP 10 EV BRANDS.png", dpi=300)
plt.title("Top 10 EV Brands")
```

```python
# Monthly EV Growth
```

```python
monthly_ev = df[df['FUEL CATEGORY'] == 'EV'].groupby(df['REGISTRATION DATE'].dt.to_period('M')).size()
monthly_ev.plot(figsize=(12,6), title="Monthly EV Growth Trend", marker='o')
plt.tight_layout()  
plt.savefig("Monthly EV Growth Trend.png", dpi=300)
```
